```python
# Determine which portfolios are riskier than the S&P 500

#Use a for loop to iterate through dataset and compare standard deviations in all other portfolios to the S&P 500
for x in Cumulative_Returns:
    if Cumulative_Returns[x].std()  > Cumulative_Returns['S&P 500'].std():
        print(f"{x} is riskier than S&P 500")
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)

#The following code is based on Tobin's formula for standard deviations which I found on Google and requires Numpy.

Variation = Cumulative_Returns.var()
Mean = Variation.mean() + 1
Annualized_std_dev = np.sqrt(((Variation + (Mean**2))**12) - Mean**24)
Annualized_std_dev
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
#I calculated the S&P variance against Berkshire Hathawaay just because cov requires 2 elements for contrast.
Covariance = Cumulative_Returns['S&P 500'].cov(Cumulative_Returns['BERKSHIRE HATHAWAY INC'])
print(f"Covariance between S&P 500 and Berkshire Hathaway: {Covariance}\n")

# Calculate variance of S&P 500
Variance = Cumulative_Returns['S&P 500'].var()
print(f"Variance of S&P 500: {Variance}\n")

# Calculate beta

Beta = Covariance / Variance

print(f"The Beta for S&P 500 is: {Beta}\n")

# Plot beta trend
#First you have to calculate the rolling covariance & variance

Rolling_Cov = Combined_Portfolio_Returns_Fixed['S&P 500'].rolling(window = 30).cov(Combined_Portfolio_Returns_Fixed['BERKSHIRE HATHAWAY INC'])
Rolling_Cov.plot()
```

```python
#Calculate variance separately
Rolling_Var = Cumulative_Returns['S&P 500'].rolling(window = 30).var()

#Plot the variance
Rolling_Var.plot()
```

```python
#Calculate the rolling beta and assign the results to its own variable
Rolling_Beta = Rolling_Cov / Rolling_Var

#Plot lot the trend, set the size of the visual and give it a title
Rolling_Beta.plot(figsize=(20, 10), title='Rolling 30-Day Beta of S&P 500') 
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
Exp_Weighted_Avg = Cumulative_Returns.ewm(halflife = 21)
Exp_Weighted_Avg.mean()
```

```python
#Plot EWM
Plot_Exp_Mv_Avg = Exp_Weighted_Avg.mean()
Plot_Exp_Mv_Avg.plot(figsize = (10,5))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
Sharpe_Ratio = (Annualized_std_dev.mean() * 252) / (Annualized_std_dev * np.sqrt(252))
Sharpe_Ratio.head()
```

```python
# Visualize the sharpe ratios as a bar plot
Sharpe_Ratio.plot.bar(title='Sharpe Ratios')
```
